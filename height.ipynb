```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataset\height_data.csv')
```

```python
sns.kdeplot(data=data, x="Height", hue="Gender", common_norm=True, fill=True)

plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.title("Height Distributions by Gender and Country")

plt.show()
```
